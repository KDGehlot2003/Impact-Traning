# Circular Linked List

- Circular linked list will we similar to single linked list but in the single linked list last node address in null nut int he circular linked list last node will be connected to first node that's why last node address is address of first node 
- Single linked list contain only head but the circular linked list contain head and tail
- Head always pointing to first node
- Tail always pointing to last node

```python
#  WAP to create 4 nodes in a circular linked list and display the data present in the linked list and 
#  reverse the data present in the node.

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class cll:
    def __init__(self):
        self.head = None
        self.tail = None
    def display(self):
        if (self.head is None):
            print("Empty")
        else:
            temp = self.head
            while(True):
                print(temp.data,end="---->")
                temp = temp.next
                if(temp == self.head):
                    break
            print(temp.data)
                
n1 = Node(10)
l1 = cll()
l1.head = n1
l1.tail = n1
l1.tail.next = l1.head
n2 = Node(20)
l1.tail.next = n2
l1.tail = n2
l1.tail.next = l1.head
n3 = Node(30)
l1.tail.next = n3
l1.tail = n3
l1.tail.next = l1.head
n4 = Node(40)
l1.tail.next = n4
l1.tail = n4
l1.tail.next = l1.head

l1.display()


```
