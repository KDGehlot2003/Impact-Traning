# Multi-Threading
- There is a functionality or logic with execute simultaneously along with the other part of the program.
- It is lightweight process.
- A program which is under execution is known as process.
- We can define functionality or logic as a thread by overriding *run* method of thread class.
- Thread class is pre-defiend class which is defined in threading module.
- Thread can be executed through __*start()*__ method of thread class.

```python
import threading

class x(threading.Thread):
    def run(self):
        for p in range(1,101):
            print(p)

class y(threading.Thread):
    def run(self):
        for q in range(101,201):
            print(q)

x1=x()
x1.start()
y1=y()
y1.start()
for r in range(201,301):
    print(r)
for i in range(301,401):
    print(i)
```

# Thread Name
- Python interpreter assign one unique name to each and every thread
- we can get names of thread by calling __*getname()*__ method of thread class.
- we can also assign our own names to thread by calling __*setname()*__ method of thread class

```python
#getName() and setName() is deprecated use .name instead

import threading

class x(threading.Thread):
    def run(self):
        print(self.name)
class y(threading.Thread):
    def run(self):
        print(self.name)
        
x1=x()
x1.name = "X-Thread"
x1.start()
y1=y()
y1.name = "Y-Thread"
y1.start()
```

# Suspending the Execution of Thread
- We can Suspend the execution of thread temporary by calling __*sleep()*__ function or __*join()*__ method
- __*sleep()*__ function is pre-defined which is defined in time module.
- __*sleep()*__ function suspend the exexution of current thread until the specified time is over.
- __*join()*__ method is pre-defined method which is defined in thread class.
- __*join()*__ method stop the execution of the current thread until specified thread execution is over.

```python
# sleep()
import time
for p in range(1,21):
    print(p)
    time.sleep(.1)
```

# Regular Expression

- If you want to represent a group of string according to the particular format or pattern then we should go for regular expression.
- Regular Expression is a declaration of machenism to reprensent a group of srting according to the particular format or pattern.


## Example
- We can write regular expression to represent mobile numbers
- We can write regular expression to represent all mail ids.
                           
## The main imp area of regular expression application:

1. To develop validation framework or validation logic
2. To develop pattern matching application 
    __Example - crtl f in windows and grep in unix__
    
3. To develop translated site compiler, interpreter and etc.
4. To develop digital circuits
5. To develop communication protocols like TCP, UDP, etc.


- We can develop re based on the by using python model regular expression
- This module contains several inbuilt function to use regular expression

## Compile

- re model contain comiple function, the compiler pattern into regex object 
- __*finditer()*__ in iteration object which is match object for every match.
- on match object we can call the following methods:
    1. start
    2. end
    3. group
    
## Important function of re module

1. match()
2. fullmatch()
3. search()
4. findall()
5. finditer()
6. sub()
7. subn()
8. split()


### 1. match()
- We can use __*match()*__ function to check the given pattern at the beggining of the target string.
- If the match is available then we will get match object otehrwise we will get None.

### 2. fullmatch()
- We can use __*fullmatch()*__ method to match a pattern to all of target string, ie complete string should be matched according to the given pattern.
- If the complete stirng match then this function will return match objectotherwise it will return None.

### 3. search()
- We can use __*search()*__ function to search to given pattern in the target string
- If match is available then it return the match object which represent the first occurance of match
- If the match is not available then it return None.

### 4. findall()
- We can use __*findall()*__ function to find all the occuurace of the match.
- This function returns a list object which contains all occuraces.

### 6. sub()
- __*sub()*__ means subsitution or replacement.
- In the target string every match pattern will be replaced with the provided replacement.

### 7. subn()
- It is exactly same as __*sub()*__ but it can also return no. of replacement


###  ^ Symbol
- We can use __^__ symbol to check whether given target string starts with our provided or not.
- If the target string starts with the given pattern, then it returns matched object, otherwise None.

###   $$ Symbol 
- We can use __$__ symbol to check whether target string end with the provided pattern or not.

```python
# finditer()
import re
c = 0
p = re.finditer("ab","abaabab")

for s in p:
    c +=1
    print(s.start(),"---",s.end(),"---",s.group())
print("No. of occurences: ",c)
```

# Pre defined Character classes:

- \s → Space character
- \S → Any character except space character 
- \d → Any digit from 0 to 9
- \D → Any character except digit 
- \w → Any word character [a-zA-Z0-9]
- \W → Any character except word character (Special Characters)
- . → Any character including special characters

```python
import re
l = re.finditer('\s','a7b @k9z')
for p in l:
    print(p.start(),"---",p.group())
```

```python
# WAP to check the given mobile no. is valid mobile no. or not.

import re
n = input("Enter mobile number")
v= re.fullmatch("[6-9]\d{9}",n)
if (v!=None):
    print("Valid mobile number")
else:
    print("Invalid mobile number")
```
