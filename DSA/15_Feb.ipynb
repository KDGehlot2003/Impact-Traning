```python
# WAP to create two single linked list with four nodes and 
# compare the element of both linked list if the element are same then return true or else return false
# WAP to match the two sorted linked list



class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class linkedlist:
    def __init__(self):
        self.head = None
        
    def display(self):
        temp = self.head
        if self.head is None:
            print("Empty hai ðŸ¥²")
        else:
            while(temp):
                print(temp.data,end="--->")
                temp = temp.next
            
def compare(s1,s2):
    c1 = s1.head
    c2 = s2.head
    while (c1 and c2):
        if c1.data != c2.data:
            return False
        c1 = c1.next
        c2 = c2.next
    
    if (c1 is None and c2 is None):
        return True
    return False
        

def SORT(l1,l2):
    temp1 = l1.head
    temp2 = l2.head
    arr = []
    while temp1 and temp2:
        if temp1.data < temp2.data:
            arr.append(temp1.data)
            temp1 = temp1.next
        else:
            arr.append(temp2.data)
            temp2 = temp2.next
    while temp1:
        arr.append(temp1.data)
        temp1 = temp1.next
    while temp2:
        arr.append(temp2.data)
        temp2 = temp2.next
    
    
    for i in arr:
        print(i,end="--->")

n1 = Node(10)
n2 = Node(30)
n1.next = n2
n3 = Node(50)
n2.next = n3
n4 = Node(70)
n3.next = n4

l1 = linkedlist()
l1.head = n1
l1.display()
print()

d1 = Node(20)
d2 = Node(40)
d1.next = d2
d3 = Node(60)
d2.next = d3
d4 = Node(80)
d3.next = d4
l2 = linkedlist()
l2.head = d1
l2.display()
print()

print(compare(l1,l2))

print()

SORT(l1,l2)
```
